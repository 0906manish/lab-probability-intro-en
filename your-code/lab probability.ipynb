```python
relativefrequency_category = df.category.value_counts(normalize= True)
relativefrequency_category
relativefrequency_categorytoplot = relativefrequency_category[0:5]
relativefrequency_categorytoplot 
```

```python
frequency_category= df.category.value_counts()
frequency_category
```

```python
frequency_category.info()
```

```python
df_frequencytable = pd.concat([frequency_category, relativefrequency_category], axis = 1 )
df_frequencytable
df_frequencytabletoplot= df_frequencytable[0: 5]
df_frequencytabletoplot
```

```python
sns.barplot(y= "count", x = "category", data = df_frequencytabletoplot, palette = "pastel", legend = True) 
plt.xticks(rotation=45)
plt.show()
#sns.countplot(x= "count", data= df_frequencytabletoplot, palette= "viridis", legend = True)
```

Since the category column has lot of entries/indices plotting a barplot or the count plot is very complicated to visualize or understand. Thus we break our df_frequencytable dataframe into chunks and plot them all together.

```python
#lista = [df_frequencytable.index]

lista = []
for index in df_frequencytable.index:
    lista.append(index)
len(lista)
listatobreak = [lista[36], lista[73],lista[110], lista[147],lista[184], lista[211], lista[248], lista[295]]
listatobreak
for i in range(len(listatobreak)):
    plt.figure(figsize=(10, 4)) #create a new plot each time
    df_i = df_frequencytable.loc[listatobreak[i]:listatobreak[i+1]]
    sns.barplot(y = "count", x = "category",data = df_i,  palette = "viridis", legend = True)
    plt.title(f"Plot for segment {i}: {listatobreak[i]} to {listatobreak[i+1]}")
    plt.xticks(rotation = 90)
    plt.show()

#for i in range(len(listatobreak)-1):
#    sns.barplot(y = "count", x = df_frequencytable.loc[listatobreak[i]:listatobreak[i+1]], data = df_frequencytable.loc[listatobreak[i]:listatobreak[i+1]], palette = "pastel", legend = True)
    
#plt.tight_layout()
#plt.show()    
    
```
